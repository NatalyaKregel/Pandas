## **Задача 1 (ДЗ)**
Дан файл california_housing_train.csv. Определить среднюю стоимость дома , где количество людей от 0 до 500 (population) и сохранить ее в переменную avg.
Используйте модуль pandas.

```python
import pandas as pd
```

```python
df = pd.read_csv('sample_data/california_housing_train.csv')
```

```python
df.head()
```

```python
# создаем маску, чтобы выбрать значение столбца population в диапозоне от 0 до 500
```

```python
mask = (df['population'] > 0) & (df['population'] < 500)
```

## фильтруем, чтоб получить нужные строки

```python
filtered_df = df[mask]
```

## использовали функцию mean, для расчета среднего значения столбца median_house_value в отсортированной таблице

```python
avg = filtered_df['median_house_value'].mean()
```

```python
df = pd.read_csv('sample_data/california_housing_train.csv')
```
